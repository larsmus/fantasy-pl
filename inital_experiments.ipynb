```python
import requests
import matplotlib.pyplot as plt
import seaborn as srs 
from constants import API_URLS, positions
```

```python
base_json = requests.get(API_URLS["static"]).json()
players = base_json["elements"]
```

```python
for p in players:
    p["name"] = p["first_name"] + " " +p["second_name"]
```

```python
costs = [p["now_cost"] for p in players]
name = [p["name"] for p in players]
points = [p["total_points"] for p in players]
player_type = [positions[p["element_type"]] for p in players]
```

```python
plt.figure(figsize=(8,6))
srs.scatterplot(x=costs, y=points, hue=player_type)
plt.show()
```
